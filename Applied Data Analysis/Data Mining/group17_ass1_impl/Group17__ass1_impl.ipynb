We can see from above transformation that we have dealt with the skewness of variables 'Revenue' and 'Amount'. The reason we take log is to transform the models into normal distributions that MAchine Learning models can actually use while reducing the impact of outliers and improving model performance.These transformations so far are for right skewed variables. Below we will look at left skew.

```python
df['Log_Return'] = np.log(df['Return'] - df['Return'].min() + 1)
df['Log_Total_Equity'] = np.log(df['Total_Equity'] - df['Total_Equity'].min() + 1)
df['Log_Operating_Cash_Flow'] = np.log(df['Operating_Cash_Flow'] - df['Operating_Cash_Flow'].min() + 1)
df['Log_Investing_Cash_Flow'] = np.log(df['Investing_Cash_Flow'] - df['Investing_Cash_Flow'].min() + 1)
df['Log_Financing_Cash_Flow'] = np.log(df['Financing_Cash_Flow'] - df['Financing_Cash_Flow'].min() + 1)
```

The log shift helps us deal with negative values and helps transform data to normal distribution to be picked up by the machine learning model.

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

# Example: Log transform skewed features
df['Log_Volume'] = np.log(df['Volume'] + 1)
df['Log_Amount'] = np.log(df['Amount'] + 1)

# Standardize numeric features
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
```

## Feature Selection

```python
# Calculate correlation matrix
corr_matrix = df.corr()

# Select upper triangle of the correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Find features with correlation greater than 0.9 (or another threshold you find suitable)
to_drop = [column for column in upper.columns if any(upper[column] > 0.9)]

# Drop those features
df_reduced = df.drop(columns=to_drop)

print("Dropped columns due to high correlation:", to_drop)
```

We have eliminated multicolinearity by dropping columns with high correlation with them.

## Recursive Feature Elimination (RFE)

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
import numpy as np

# Ensure that 'Date' column is dropped, but keep 'Year', 'Month', and 'Quarter'
X = df_reduced.drop(columns=['Volatility', 'Date'], errors='ignore')

# Convert boolean columns to numeric (0/1)
X = X.applymap(lambda x: 1 if x is True else 0 if x is False else x)

# Check if X contains only numeric data
print("Columns in X (should be all numeric):\n", X.dtypes)

# If there are still any non-numeric columns, drop them
non_numeric_cols = X.select_dtypes(exclude=[np.number]).columns
if len(non_numeric_cols) > 0:
    print(f"Non-numeric columns found and will be removed: {non_numeric_cols}")
    X = X.drop(columns=non_numeric_cols)

# Check for missing values in X
print(f"Number of missing values in X: {X.isna().sum().sum()}")

# Impute missing values in X with the mean
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Check for missing values in target variable y
y = df_reduced['Volatility']
print(f"Number of missing values in y before handling: {y.isna().sum()}")

# Handle missing values in y (either drop them or fill with a value, here we drop)
y = y.dropna()

# Ensure that corresponding rows in X are also dropped to match y
X_imputed = X_imputed[:len(y)]

print(f"Shape of X_imputed: {X_imputed.shape}")
print(f"Shape of y: {y.shape}")

# Now apply RFE to select top features (optimized by using step=5 to reduce number of iterations)
model = LinearRegression()
rfe = RFE(estimator=model, n_features_to_select=10, step=5)  # Eliminate 5 features at a time
fit = rfe.fit(X_imputed, y)

# Get the selected features
selected_features_rfe = X.columns[fit.support_]

print("Selected Features by RFE:\n", selected_features_rfe)
```

# Final Model with best output

## Lasso Regression

```python
from sklearn.model_selection import GridSearchCV, KFold, train_test_split  # Import train_test_split
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
import numpy as np

# Define X (features) and y (target variable)
X = df_reduced.drop(columns=['Volatility','Date'])
y = df_reduced['Volatility']

# Impute missing values in X with the mean
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Define the hyperparameter grid for tuning alpha
param_grid = {
    'alpha': np.logspace(-4, 1, 50)  # Trying alpha values from 1e-4 to 10 on a log scale
}

# Initialize the Lasso model
lasso = Lasso(max_iter=10000)

# Set up cross-validation with 5 folds
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Use GridSearchCV to find the best alpha
grid_search = GridSearchCV(lasso, param_grid, cv=kf, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search.fit(X_train_scaled, y_train)

# Get the best hyperparameters
best_alpha = grid_search.best_params_['alpha']
print(f"Best alpha found: {best_alpha}")

# Train the Lasso model with the best alpha
lasso_best = Lasso(alpha=best_alpha, max_iter=10000)
lasso_best.fit(X_train_scaled, y_train)

# Predict on the test set
y_pred = lasso_best.predict(X_test_scaled)

# Calculate RMSE and R-squared on the test set
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

# Output the final results
print(f"RMSE on the test set: {rmse}")
print(f"R-squared on the test set: {r2}")
```

As we can see from above output, a small Alpha value indicates less regularization was needed which means the model fits the data well. The RMSE score is average (but best performing among all models) and shows that the predicted volatilities differ from actual values by 0.54 units. Finally, the R squared shows us that it explains about 54% of the variance.

```python
# Load the submission data for predictions
submission_df = pd.read_csv('A1_stock_volatility_submission.csv')

stock_column = submission_df['Stock']  # Retain the original Stock column for the output

# Apply the same preprocessing as done on the training data
submission_df = pd.get_dummies(submission_df, columns=['Stock'], drop_first=True)

# Handle missing features by adding them with default values (e.g., 0)
missing_features = [feature for feature in X.columns if feature not in submission_df.columns]
for feature in missing_features:
    submission_df[feature] = 0

# Apply log transformations if needed
if 'Volume' in submission_df.columns and 'Log_Volume' not in submission_df.columns:
    submission_df['Log_Volume'] = np.log(submission_df['Volume'] + 1)

if 'Amount' in submission_df.columns and 'Log_Amount' not in submission_df.columns:
    submission_df['Log_Amount'] = np.log(submission_df['Amount'] + 1)

# Ensure the submission data has the same features as the training data
submission_df_final = submission_df[X.columns]

# Standardize the submission data using the same scaler applied to training data
submission_df_scaled = scaler.transform(submission_df_final)

# Predict the volatility on the submission data using the fine-tuned Lasso model
predicted_volatility = lasso_best.predict(submission_df_scaled)

# Create the output DataFrame
output_df = pd.DataFrame({

    'Stock': stock_column,  # Keeping the original Stock column
    'Predicted Volatility': predicted_volatility
})

# Save the predictions to a CSV file
output_file_path = 'pred_values.csv'
output_df.to_csv(output_file_path, index=False)

print("Predicted values saved to:", output_file_path)
```
